```python
import torch
```

```python
torch.manual_seed(0)
torch.rand(2, 10, 5, 5).permute(0, 2, 3, 1)
```

```python
embedding = torch.rand(5, 10)
z = torch.rand(2, 10)
```

```python
torch.sum(z**2, dim=-1, keepdim=True).shape
```

```python
torch.sum(embedding**2, dim=-1).shape
```

```python
torch.sum(z**2, dim=-1, keepdim=True) + torch.sum(embedding**2, dim=-1)
```

```python
2*(torch.matmul(z, embedding.T))
```

```python
euclidia_dis = torch.sum(z**2, dim=-1, keepdim=True) + torch.sum(embedding**2, dim=-1) - 2*(torch.matmul(z, embedding.T))
euclidia_dis
## Here the first value tells the euclidean distance between first latent z[0] and first embedding, then first latent -> to second embeddng , then third and so forth
## next row saying , second latent dim z[1] to first, second, third and so forth euclidian distance with embeedding
```

```python
# next we take lowest embedding for each z, by euclidean distance
min_euc_indieces = euclidia_dis.argmin(dim=-1)
```

```python
# next take the embedding 
embedding[min_euc_indieces]
# the first embedding below is the embedding that is close to first z, the second embedding below for the second z
```
